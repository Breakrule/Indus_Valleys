```python
import cv2

ini = cv2.imread("crop_5/0.png")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(100, 100))
plt.imshow(ini)
```

```python
import numpy as np

haaa = ini[0,:,0]
list_garis = np.asarray([_ for _ in range(len(haaa))])[haaa!=255]
hhhh = [0]
for garis in list_garis:
    hhhh.append(garis)
list_garis = np.asarray(hhhh)
```

```python
list_garis
```

```python
for i in range(len(list_garis)-1):
    try:
        cv2.imwrite("crop_5/"+str(i)+".png", ini[:, list_garis[i]+1: list_garis[i+1]])
        #print(i)
    except:
        pass
```
